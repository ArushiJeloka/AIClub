# Data Types, Operators, and Print Statements
A data type is an attribute of data which tells the interpreter how the data is intended to be used. In Python, the most commonly used datatypes are numeric, strings, booleans, lists, tuples, sets, and dictionaries. In this lesson, we will go over numerics, strings, booleans, operators and print statements.

```python
# To check the datatype, use the type() function
# To print an output use the print() function
```

```python
import math
#Basic Binary Operators (need at least 2 numbers): +, -, *, /, **, %, //
print(1 + 1) # addition
print (1 - 2) # subtraction
print(88 * 79) # multiplication
print(4 / 3) # division
print(8 % 5) # modulus - gives remainder
print(2**3) # exponentiation
print(4//3) # floor division
print(math.ceil(4/3)) # ceiling division
```

```python
#The `-` or '+' operators can also be used as unary operators (operate on only one number)
print(5+-10)
print(-5 +- 3)
print(5-+8)
```
