```python
import seaborn as sns 
%matplotlib inline
```

```python
tip = sns.load_dataset('tips')
tip.head()
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tip)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tip, hue = 'time')
```

```python
sns.countplot(x = 'time', data = tip)
```

```python
sns.barplot(x = 'time', y = 'total_bill', data = tip)
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tip)
```

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = tip)
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tip)
```

```python
sns.distplot(tip['total_bill'])
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tip)
```

```python
sns.pairplot(tip)
```
